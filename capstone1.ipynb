# The dataframe has one point, two point, three point, four point and more than four point mutations. Counted the number 
# of each mutations in the data set using the function mutation_counter and created a list for each type and for sorting the
# mutation as a 1,2,3,4 and >4 

```python
def mutation_counter(col):
    ### This function counts the number of mutations in each entry of the dataframe column 'mutations' ###
    ### and returns the length of each type of mutation in 4 different lists ('one','two','three','four','more') ###
    ### and a fifth list, 'count' which contains the designation for each entry as a 1,2,3,4 or >4. ###
    
    one = []                                                           # Created empty lists for each type
    two = []
    three = []
    four = []
    more = []
    count = []                                                          # list for designating the type of mutation
    
    for i in col:                                                       # Iterating through the different entries in the column
        c = i.count('_')                                                # c is the number of '_' in the entry
        if c == 0:                                                      # if c is 0, its a one point mutation
            one.append(i)                                               # added to the list 'one'
            count.append('1')                                           # designated as a 1 in the count list
        elif c == 1:                                                    # if c is 1, its a two point mutation
            two.append(i)                                               # added to list 'two'
            count.append('2')                                           # designated as a 2 in the count list
        elif c == 2:                                                    # if c is 2, its a three point mutation
            three.append(i)                                             # added to list 'three'
            count.append('3')                                           # designated as a 3 in the count list
        elif c == 3:                                                    # if c is 3, its a four point mutation
            four.append(i)                                              # added to list 'four'
            count.append('4')                                           # designated as a 4 in the count list
        else:
            more.append(i)                                              # anything more than 4 gets appended to 'more'
            count.append('>4')                                          # designated as '>4' in the count list
            
    return(len(one), len(two), len(three), len(four), len(more), count)        # return the length of each list
```

```python
# Count the number of each mutation in the dataframe df
one, two, three, four, more, count = mutation_counter(df['mutations'])
```

```python
# Selected only the entries which have one or two point mutations, ie., count is either '1' or '2' and subset in dataframe
# df_selected

df_selected = df[(df['count'] == '1') | (df['count'] == '2')]
df_selected.head(10)
```

```python
# Adding the distance feature of the mutations
# For the one point mutation, the distance is 0
# for the 2 point mutation, the position of the 2 point where the mutation occurred is taken and the absolute value of their
# difference is the dist

dist = []                                                            # empty list 'dist'
for i in df['mutations']:                                            # iterating through each entry in 'mutations' column
    c = i.count('_')                                                 # c is the count of '_'
    if c == 0:                                                       # if its 0, then 1 point mutation
        dist.append(0)                                               # and dist is appended as 0 for that entry
    if c == 1:                                                       # If c is 1, then it is a 2 point mutation
        a, b = i.strip('%').split('_')                               # strip off '%' and split at '_' and get 2 strings - a & b
        match_a = re.match(r"([a-z]+)([0-9]+)([a-z]+)", a, re.I)     # re for a - alphabet/number/alphabet
        match_b = re.match(r"([a-z]+)([0-9]+)([a-z]+)", b, re.I)     # re for b - alphabet/number/alphabet
        if match_a:
            items_a = match_a.groups()                               # group the 3 diff components into items_a
            pos1 = int(items_a[1])                                   # the 2nd component of items_a is the position of the aminoacid
        if match_b:    
            items_b = match_b.groups()                               # group the 3 diff components into items_b
            pos2 = int(items_b[1])                                   # the 2nd component of items_b is the position of the aminoacid
        dist.append(abs(pos1 - pos2))                                # absolute difference of the positions is the distance
len(dist)                                                            # check length of the 'dist' to see if it matches the 
                                                                     # number of rows in the selected dataframe, df_selected
```

```python
df_selected['mutation_distance'] = dist
```
